```python
data_copy=data_copy.drop(columns=['Sales Last Year'])
```

```python
#Adding Sales last year Column
SLY=[]
first=True
c=0
for i in data_copy['Year']:
    if i==2010:
        SLY.append(0)
        if first==False:
            c=c+1
    elif i>2010:
        SLY.append((data[data_copy['Year']==i-1]['Sales'].values[c]))
        first=False
```

```python
data['Sales Last Year']=SLY
```

```python
data.head(15)
```

```python
#Formula for Sales Growth % = (Sales in current year/Sales in Last year)-1
#Adding Sales Growth Column
SG=[]
first=True
x=0
counter=1
for i in data['Year']:
    if counter==10:
        counter=0
        x=x+1
    if i==2010:
        SG.append(0)
        counter=+1
    elif i>2010:
        value=np.round((((data[data['Year']==i]['Sales'].values[x])/(data[data['Year']==i]['Sales Last Year'].values[x]))-1)*100,2)
        SG.append(value)
        counter=counter+1

```

```python
data['Sales Growth %']=SG
```
